```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv('data/showwcase_sessions.csv')
```

```python
df.head()
```

```python
# drop bottom two rows with all null values 
df.dropna(axis=0, how='all', inplace=True)
```

```python
df.info()
```

```python
# there is one null value in sessions_likes_given column. Replacing it with 0
df['session_likes_given'] = df['session_likes_given'].fillna(0)
```

```python
# converting login_date from string to datetime so I can do feature engineering and extract day of week 
df['login_date'] = pd.to_datetime(df['login_date'])
df['day_of_week'] = df['login_date'].dt.dayofweek
```

```python
df.head()
```

```python
df.describe()
```

## EDA

```python
total = df['customer_id'].count()
unique = df['customer_id'].nunique()
print('There were total {} sessions that occurred in October 2019.'.format(total))
print('There were {} unique users.'.format(unique))
```

```python
session_mean = df['session_duration'].mean()
above_mean = df[(df['session_duration'] > session_mean)].shape[0]
below_mean = df[(df['session_duration'] < session_mean)].shape[0]
print('There are {} sessions with a duration above the session duration mean.'.format(above_mean))
print('There are {} sessions with a duration below the session duration mean.'.format(below_mean))
```

### How many logins were there per day in October?

```python
plt.figure(figsize=(14,6))
df.set_index('login_date').session_id.resample('D').count().plot()
plt.xlim('2019-10-01', '2019-11-01') 
plt.xlabel('Login Date', fontsize=16)
plt.ylabel('Number of Logins', fontsize=16)
plt.title('Number of Logins for October 2019', fontsize=18)
plt.show()
```

### What day of the week generated the most traffic?

```python
x = df['day_of_week'].value_counts()
labels = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
index = np.arange(len(x))

plt.figure(figsize=(6,4))
plt.bar(x.index, x.values, color='lightcoral')
plt.xticks(index, labels, fontsize=10, rotation=0)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Number of Logins by Day of Week', fontsize=16)
plt.show()
```

### How did user activity vary between day of week?

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(22,5))

df.groupby('day_of_week').session_projects_added.mean().plot(kind='bar', color='darkgreen', alpha=0.7, edgecolor='g', ax=ax1, rot=0)
ax1.set_title('Average Projects Added by Day of Week', fontsize=18)
ax1.set_ylabel("Average Count")

df.groupby('day_of_week').session_likes_given.mean().plot(kind='bar', color='darkgreen', alpha=0.7, edgecolor='g',ax=ax2, rot=0)
ax2.set_title('Average Likes Given by Day of Week', fontsize=18)
ax2.set_ylabel("Average Count")

df.groupby('day_of_week').session_comments_given.mean().plot(kind='bar', color='darkgreen', alpha=0.7, edgecolor='g', ax=ax3, rot=0)
ax3.set_title('Average Comments Given by Day of Week', fontsize=18)
ax3.set_ylabel("Average Count")
```

```python
a = df.groupby('day_of_week').session_projects_added.sum()
b = df.groupby('day_of_week').session_likes_given.sum()
c = df.groupby('day_of_week').session_comments_given.sum()

labels = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
index = np.arange(len(a))
colors = ['lightblue', '#00BFFF', 'darkgreen']
x = df['day_of_week'].value_counts()
r = range(len(x))

barWidth = 0.8
plt.figure(figsize=(8,6))
plt.bar(r, a, color=colors[0],edgecolor='white', width=barWidth, label="Projects Added")
plt.bar(r, b, bottom=np.array(a), color=colors[1], edgecolor='white', width=barWidth, label='Likes Given')
plt.bar(r, c, bottom=np.array(a)+np.array(b), color=colors[2], alpha=0.7, edgecolor='white', width=barWidth, label='Comments Given')
plt.xticks(index, labels, fontsize=10, rotation=0)
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.title('Total User Activity by Day of Week', fontsize=16)
plt.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(22,5))

df.groupby('day_of_week').inactive_duration.mean().plot(kind='bar', color='darkorange', ax=ax1, rot=0)
ax1.set_title('Average Inactive Duration by Day of Week', fontsize=18)
ax1.set_ylabel("Average Count")

df.groupby('day_of_week').bugs_in_session.mean().plot(kind='bar', color='darkorange', ax=ax2, rot=0)
ax2.set_title('Average Bugs in Session by Day of Week', fontsize=18)
ax2.set_ylabel("Average Count")

df.groupby('day_of_week').session_duration.mean().plot(kind='bar', color='darkorange', ax=ax3, rot=0)
ax3.set_title('Average Session Duration by Day of Week', fontsize=18)
ax3.set_ylabel("Average Count")
```

```python
d = df.groupby('day_of_week').inactive_duration.sum()
e = df.groupby('day_of_week').session_duration.sum()

labels = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
index = np.arange(len(d))
colors = ['lightblue', 'darkorange']
x = df['day_of_week'].value_counts()
r = range(len(x))

barWidth = 0.8
plt.figure(figsize=(8,6))
plt.bar(r, d, color=colors[0], edgecolor='white', width=barWidth, label="Inactive Duration")
plt.bar(r, e, bottom=np.array(d), color=colors[1], edgecolor='white', width=barWidth, label='Session Duration')
plt.xticks(index, labels, fontsize=10, rotation=0)
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.title('Active vs. Inactive Session by Day of Week', fontsize=16)
plt.legend()
```

### Next I to know how many repeated users there were, and what did they mostly do? 

```python
# find repeat users using groupby and filter out count > 1
repeat = df[['customer_id','session_id']].groupby('customer_id').count()
repeat = repeat[repeat['session_id']>1]
repeat.shape
```

```python
group = df[['customer_id','session_projects_added', 'session_likes_given', 'session_comments_given', 
            'inactive_duration', 'bugs_in_session', 'session_duration']].groupby('customer_id').sum()
group.head()
```

```python
merged = pd.merge(repeat, group, left_index=True, right_index=True)
merged
```

```python
merged.describe()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(22,5))

j = ax1.hist(merged.session_projects_added, bins=20, color='red', alpha=.6)
k = ax2.hist(merged.session_likes_given, bins=20, color='red', alpha=.6)
l = ax3.hist(merged.session_comments_given, bins=20, color='red', alpha=.6)
ax1.set_title('Projects Added Distribution', fontsize=18)
ax2.set_title('Likes Given Distribution', fontsize=18)
ax3.set_title('Comments Given Distribution', fontsize=18)
ax1.set_xlabel("Number of Projects", fontsize=14)
ax2.set_xlabel("Number of Likes Given", fontsize=14)
ax3.set_xlabel("Number of Comments Given", fontsize=14)
ax1.set_ylabel("Count")
ax2.set_ylabel("Count")
ax3.set_ylabel("Count")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(22,5))

j = ax1.hist(group.session_projects_added, bins=20, color='red', alpha=.6)
k = ax2.hist(group.session_likes_given, bins=20, color='red', alpha=.6)
l = ax3.hist(group.session_comments_given, bins=20, color='red', alpha=.6)
ax1.set_title('Projects Added Distribution', fontsize=18)
ax2.set_title('Likes Given Distribution', fontsize=18)
ax3.set_title('Comments Given Distribution', fontsize=18)
ax1.set_xlabel("Number of Projects", fontsize=14)
ax2.set_xlabel("Number of Likes Given", fontsize=14)
ax3.set_xlabel("Number of Comments Given", fontsize=14)
ax1.set_ylabel("Count")
ax2.set_ylabel("Count")
ax3.set_ylabel("Count")
```
